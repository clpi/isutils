```python
from models.demo.demo import Demo
import lxml.etree as ET
import copy
import uuid
```

```python
demo = r"C:\Users\Jess\...T\imp\cg1.demo"
script = r"C:\Users\Jess\...T\imp\85- CodeGuru Review for GitHub Enterprise [R1 - V2] - ready to import.docx"
```

```python
d = Demo(path=demo, script_path=script)
```

```python
root = d.root
root_c = copy.deepcopy(d.root)
for i, sect in enumerate(root.findall("Chapters/Chapter")):
    for j, step in enumerate(sect.findall("Steps/Step")):
        pass

def clear_sects(root):
    root.find("Chapters".clear())
    return root
    
def insert_sect(root):
    chapters = root.find("Chapters")
    num_chapters = len(chapters.findall("Chapter"))
    chapter = ET.SubElement(chapters, "Chapter")
    id_el = ET.SubElement(chapter, "ID")
    id_el.text = str(uuid.uuid4())
    xml_name_el = ET.SubElement(chapter, "XmlName")
    name_el = ET.SubElement(xml_name_el, "Name")
    name_el.text = r"<![CDATA[Section "+str(num_chapters+1)+r"]]>"
    steps = ET.SubElement(chapter, "Steps")
    is_active_el = ET.SubElement(chapter, "IsActive")
    is_active_el.text = "true"
    click_anywhere_el = ET.SubElement(chapter, "ClickAnywhere")
    click_anywhere_el.text = "false"
    
def append_step(chapter, step):
    steps_el = chapter.find("Steps")
    steps_el.append(step)
```

```python
print(root_c.findall("Chapters/Chapter").__len__())
root_c.find("Chapters").clear()
print(root_c.findall("Chapters/Chapter").__len__())
print(ET.tostring(root_c, pretty_print=True))
insert_sect(root_c)
insert_sect(root_c)
print(ET.tostring(root_c, pretty_print=True))
print(root_c.findall("Chapters/Chapter").__len__())
og_step1 = root.findall("Chapters/Chapter")[0].findall("Steps/Step")[0]
append_step(root_c.findall("Chapters/Chapter")[0], og_step1)
print(ET.tostring(root_c))
```
